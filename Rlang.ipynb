mpg : 엔진 크기와 연비와의 관계

hwy: 고속도로 연비   


````r
mpg
````

````r
#산점도
ggplot(data=mpg)+
  geom_point(mapping = aes(x=displ, y=hwy))
````

````r
#ggplot은 graph를 그릴 수 있는 자리를 제공
ggplot(data=mpg)
````

심미성 매핑 -> 시각적 속성: 점의 크기, 모양, 색상  
class: 소형, 중형, SUV 등  
심미성 변수 매핑을 위해 aes 내부에 인수추가  

````r
ggplot(data=mpg) +
  geom_point(mapping = aes(x=displ, y=hwy, size=class))
````

````r
#알파(alpha): 투명도
ggplot(data=mpg) +
  geom_point(mapping = 
    aes(x=displ, y=hwy, alpha=class))
````

````r
#shape: 점의 모양
#주의: 6개만 제공
ggplot(data=mpg) +
  geom_point(mapping = 
    aes(x=displ, y=hwy, shape=class))
````

````r
#color: 색상
ggplot(data=mpg) +
  geom_point(mapping = 
    aes(x=displ, y=hwy, color=class))
````

````r
#color: 색상 수동 맵핑
#aes 밖에 지정
ggplot(data=mpg) +
  geom_point(mapping = 
    aes(x=displ, y=hwy),color="blue")
````

````r
#면분할(facet)
ggplot(data=mpg) +
  geom_point(mapping = aes(x=displ, y=hwy)) +
  facet_wrap(~class, nrow = 2)
````

````r
#두 변수 조합 면분할 : faect_grid
ggplot(data=mpg) +
  geom_point(mapping = aes(x=displ, y=hwy)) +
  facet_grid(drv ~ cyl)
````

````r
#geom_smooth(음영은 오차범위)
ggplot(data=mpg) +
  geom_smooth(mapping = aes(x=displ, y=hwy))

#drive 진진,후진 사륜 구동에 따른 그래프 그리기
ggplot(data=mpg) +
  geom_smooth(mapping = aes(x=displ, y=hwy, group=drv))
ggplot(data=mpg) +
  geom_smooth(mapping = aes(x=displ, y=hwy, color=drv))

#범례(legend) 삭제
ggplot(data=mpg) +
  geom_smooth(mapping = aes(x=displ, y=hwy, color=drv), show.legend = F)
````

````r
#여러 플롯 그리기

ggplot(data=mpg)+
  geom_point(mapping = aes(x=displ, y=hwy)) +
  geom_smooth(mapping = aes(x=displ, y=hwy))

ggplot(data=mpg, mapping = aes(x=displ, y=hwy)) +
  geom_point(mapping = aes(color=class)) +  
  geom_smooth()
````

````r
#filter

ggplot(data=mpg, mapping = aes(x=displ, y=hwy)) +
  geom_point(mapping = aes(color=class)) +  
  geom_smooth(data=filter(mpg, class=="subcompact"))
````
